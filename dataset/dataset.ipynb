```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import os
import openslide
import xml.etree.cElementTree as ET
import random
import os
import glob
import re

path = '/content/colab_directory/CRC_WSIs_original/'
```

```python
# get all file names from folder
svs_names = []
xml_names = []
for file_name in os.listdir(path):
  if file_name.endswith('svs'):
    svs_names.append(file_name)
  elif file_name.endswith('xml'):
    xml_names.append(file_name)

svs_names = sorted(svs_names, key=lambda x: int(re.search(r'\d+', x).group()))
xml_names = sorted(xml_names, key=lambda x: int(re.search(r'\d+', x).group()))
```

```python
def parse_xml(anno_path, factor):
    tree = ET.ElementTree(file=anno_path)
    annolist = []
    root = tree.getroot()
    for coords in root.iter('Vertices'):
        vasc = []
        for coord in coords:
            x = int(float(coord.attrib.get("X")) / factor)
            y = int(float(coord.attrib.get("Y")) / factor)
            vasc.append((x, y))
        annolist.append(vasc)
    return annolist
```

```python
def point_in_polygon(x, y, poly):
    """
    Check if a point is inside a polygon using the ray casting algorithm.
    """
    num = len(poly)
    j = num - 1
    odd_nodes = False
    for i in range(num):
        if poly[i][1] < y and poly[j][1] >= y or poly[j][1] < y and poly[i][1] >= y:
            if poly[i][0] + (y - poly[i][1]) / (poly[j][1] - poly[i][1]) * (poly[j][0] - poly[i][0]) < x:
                odd_nodes = not odd_nodes
        j = i
    return odd_nodes
```

```python
def generate_patches_ROI(path, svs_name, xml_name, patch_size, mlevel):
    slide_path = os.path.join(path, svs_name)
    print(f"Slide path: {slide_path}")
    anno_path = os.path.join(path, xml_name)
    print(f"Annotation path: {anno_path}")
    Slide = openslide.OpenSlide(slide_path)
    factor = Slide.level_downsamples[mlevel]
    annolist = parse_xml(anno_path, factor)
    slide_width, slide_height = Slide.dimensions

    # Ensure the patches directory exists
    patches_dir = f'patches_{patch_size}_{mlevel}_{svs_name}_ROI'
    if not os.path.exists(patches_dir):
        os.makedirs(patches_dir)

    for i, coords in enumerate(annolist):
        coords = np.array(coords, dtype=np.int32)

        # Create a mask for the current ROI
        mask = np.zeros((Slide.level_dimensions[mlevel][1], Slide.level_dimensions[mlevel][0]), dtype=np.uint8)
        cv.fillPoly(mask, [coords], 255)

        x, y, w, h = cv.boundingRect(coords)

        # Adjust x and y to be multiples of patch_size
        x = (x // patch_size) * patch_size
        y = (y // patch_size) * patch_size

        # Adjust w and h to cover the entire ROI while being multiples of patch_size
        w = ((w + patch_size - 1) // patch_size) * patch_size
        h = ((h + patch_size - 1) // patch_size) * patch_size

        # Calculate number of patches that can fit horizontally and vertically within the bounding box
        num_patches_x = w // patch_size
        num_patches_y = h // patch_size

        for px in range(num_patches_x):
            for py in range(num_patches_y):
                spointx = x + px * patch_size
                spointy = y + py * patch_size
                spointx0, spointy0 = int(spointx * factor), int(spointy * factor)  # Convert to original coordinates

                # Define the four corners of the patch
                corners = [(spointx, spointy), (spointx + patch_size, spointy), (spointx, spointy + patch_size), (spointx + patch_size, spointy + patch_size)]

                # Check if all four corners are within the ROI
                if all(point_in_polygon(c[0], c[1], coords) for c in corners):
                    patchimg = Slide.read_region((spointx0, spointy0), mlevel, (patch_size, patch_size))

                    # Convert to the right format for saving
                    patch_array = np.array(patchimg)
                    patch_filename = os.path.join(patches_dir, f"{svs_name}_patch_{i}_{spointx}_{spointy}.jpg")
                    print(f"Saving patch: {patch_filename}")
                    if not cv.imwrite(patch_filename, patch_array):
                        raise Exception(f"Could not write image {patch_filename}")
```

```python
# Example usage
path = '/content/colab_directory/CRC_WSIs_original/'

# List of SVS and XML file names
svs_names = [f"{i}.svs" for i in range(1, 25)]
xml_names = [f"{i}.xml" for i in range(1, 25)]

# svs_names = "4.svs"
# xml_names = "4.xml"
# Process each pair of SVS and XML files
for svs_name, xml_name in zip(svs_names, xml_names):
    # for mag_level in range(1):  # Loop through magnification levels 0, 1, and 2
    generate_patches_ROI(path, svs_name, xml_name, 512, 1)
```

```python
def is_not_white_or_gray(patch, threshold=0.95, gray_threshold=200):
    """
    Check if the patch is not predominantly white or light gray.
    """
    gray = cv.cvtColor(np.array(patch), cv.COLOR_RGB2GRAY)
    _, binary = cv.threshold(gray, gray_threshold, 255, cv.THRESH_BINARY)
    foreground_pixels = cv.countNonZero(binary)
    foreground_ratio = foreground_pixels / (patch.shape[0] * patch.shape[1])
    return foreground_ratio < threshold

def generate_patches(path, svs_name, patch_size, mlevel):
    slide_path = os.path.join(path, svs_name)
    Slide = openslide.OpenSlide(slide_path)
    downsampling_factor = Slide.level_downsamples[mlevel]

    patches_dir = f'patches_{patch_size}_{mlevel}_{svs_name}'
    if not os.path.exists(patches_dir):
        os.makedirs(patches_dir)

    slide_width, slide_height = Slide.level_dimensions[mlevel]

    for x in range(0, slide_width - patch_size, patch_size):
        for y in range(0, slide_height - patch_size, patch_size):
            patchimg = Slide.read_region((int(x * downsampling_factor), int(y * downsampling_factor)), 0, (patch_size, patch_size))
            patchimg_np = np.array(patchimg)
            if is_not_white_or_gray(patchimg_np):
                patch_filename = os.path.join(patches_dir, f"{svs_name}_patch_{x}_{y}.jpg")
                if not cv.imwrite(patch_filename, patchimg_np):
                    raise Exception("Could not write image")
                print(f"Saving patch: {patch_filename}")
```

```python
# Path to the directory containing the files
path = '/content/colab_directory/CRC_WSIs_original/'

# List of SVS file names
svs_names = [f"{i}.svs" for i in range(1, 25)]

# Example usage:
patch_size = 512  # Adjust patch size as needed

# svs_names = "4.svs"

# Process each pair of SVS
# for svs_name in svs_names:
    # for mag_level in range(1):  # Loop through magnification levels 0, 1, and 2
generate_patches(path, svs_names, patch_size, 1)
```
